```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import data set
price=pd.read_csv(r"C:\Users\Ali Mir\Desktop\Book123.csv")
```

```python
price
```

```python
#creating dummydata for binary system using pandas
pd.get_dummies(price.location)
```

```python
#making a new variable
dummies=pd.get_dummies(price.location)
dummies
```

```python
#now adjoing or merging with orignal data frame
merged=pd.concat([price,dummies],axis='columns')
merged
```

```python
#droping location column
new_price=merged.drop(['location'],axis=1)
new_price
```

```python
#now creat a model
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
#independent variable
x=new_price.drop('price',axis=1)
x
```

```python
#dependent variable
y=new_price.price
y
```

```python
#lets go to train our model
model.fit(x,y)
```

```python
model.predict([[3000,0,1,0,1]])
```

```python
#now predict westwood price
model.predict([[3000,0,0,1,0]])
```

```python
model.score(x,y)
```
